### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t').set_index('order_id')
```

```python
chipo
```

### Step 4. How many products cost more than $10.00?

```python
chipo['search_price'] = chipo.item_price.map(lambda p: p[1:]).astype('float')
new_chipo = chipo.drop_duplicates(subset=['item_name', 'quantity'])
new_chipo.loc[(new_chipo.quantity == 1) & (new_chipo.search_price > 10)].groupby('item_name').min().shape
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
new_chipo.loc[:, ['item_name', 'item_price']]
new_chipo.loc[(new_chipo.quantity == 1), ['item_name', 'search_price']].groupby('item_name').min().sort_values(by='search_price', ascending=False)
```

### Step 6. Sort by the name of the item

```python
chipo.sort_values(by='item_name', ascending=True)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.sort_values(by='search_price', ascending=False).head(1)
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo.loc[chipo.item_name == 'Veggie Salad Bowl'].quantity.sum()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo.loc[(chipo.item_name == 'Canned Soda') & (chipo.quantity > 1)].order_id.co
```
